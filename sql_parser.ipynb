```python
import re
from sql_metadata import Parser
from sqlglot import parse_one, exp
```

```python
mvp_bp_index_path = "./typ-entregables-tbp/mvp_bp_index.sql"

with open(mvp_bp_index_path,'r') as f:
    data_queries = f.read()
    
```

```python
regexp = "<sql>--[\S\s]*?--</sql>"
query_list = re.findall(regexp,data_queries)

def remove_leads(query):
    for char in ["<sql>--","--</sql>"]:
        query = query.replace(char,'')
    
    return query

queries = [ remove_leads(query) for query in query_list if re.search('caba-piba-raw-zone-db',query)]
```

```python
[parse_one(query).find_all(exp.ALL_FUNCTIONS) for query in querys]
    
```

```python
def get_meta_data(queries):
    for index,query in enumerate(queries,start=1):
        try:
            print(f"Proceso {index} : ")
            data = Parser(query)
            print(data.columns)
            
        except Exception as e:
            print(f"Error en la query #{index}")
            print({"\terror":e})
```

```python
get_meta_data(queries)
```
